```python
def loadData(filename):
    data=pd.read_csv(filename, header=None)
    x=np.array(data.iloc[:,1:])
    y=np.array(data.iloc[:,0])
    for i in range(len(y)):
        if y[i]==0:
            y[i]=-1
    y=y.reshape(len(y),1)
    return x,y
```

```python
data_train,label_train=loadData('train.csv')
data_test,label_test=loadData('test.csv')
regularisation_para_C=100
```

```python
def svm_train_primal ( data_train , label_train , regularisation_para_C ):
    N,M=data_train.shape
    var_w=cvx.Variable((M,1))
    var_b=cvx.Variable()
    var_k=cvx.Variable((N,1))
    obj=cvx.Minimize(0.5*cvx.norm2(var_w)**2+regularisation_para_C/N*cvx.sum(var_k))
    cons=[
        cvx.multiply(label_train,data_train*var_w+var_b)>=1-var_k,
        var_k>=0
    ]
    prob=cvx.Problem(obj,cons)
    prob.solve()
    svm_model={
        'w':var_w.value,
        'b':var_b.value
    }
    return svm_model
```

```python
def svm_predict_primal ( data_test , label_test , svm_model ):
    w=svm_model['w']
    b=svm_model['b']
    label_pred=np.dot(data_test,w)+b
    pred_true=0
    pred_total=len(label_test)
    for i in range(pred_total):
        if label_pred[i][0]*label_test[i][0]>0:
            pred_true+=1
    return pred_true/pred_total
```

```python
svm_model = svm_train_primal ( data_train[:LL] , label_train[:LL] , regularisation_para_C )
test_accuracy = svm_predict_primal ( data_test , label_test , svm_model )
```

```python
sum_w=np.sum(svm_model['w'])
print('the solution of b:{:.5f}'.format(svm_model['b']))
print('sum of all dimensions of w solution:{:.5f}'.format(sum_w))
print('test accuracy of svm_model:{:.5f}'.format(test_accuracy))
```

```python
def chooseC():
    optimal_acc=-1
    optimal_C=-1
    for i in range(-10,11):
        C=2**i
        str_C='2^'+str(i)
        svm_model = svm_train_primal ( data_train , label_train , C)
        test_accuracy = svm_predict_primal ( data_test , label_test , svm_model )
        print('C:{}--accuracy:{}'.format(str_C,test_accuracy))
        if test_accuracy>optimal_acc:
            optimal_C='2^'+str(i)
            optimal_acc=test_accuracy
    return optimal_C
```

```python
optimal_C=chooseC()
print('the optimal C is ',optimal_C)
```

```python
import cvxpy as cp
import numpy as np

def svm_train_dual ( data_train , label_train , regularisation_para_C ):
    N,M=data_train.shape
    
    # y.T: 1xN NxM => 1xM    @ Mx1
    # (y.T@X) @ (y.T@X).T
    # = (y.T@X) @ (X.T @ y)
    # = y.T @ (X@X.T) @ y
    # (AB)^T = B^T A^T
    # A@B@C = A@(B@C) 
    X, y = data_train, label_train
    
    A=np.multiply(np.dot(label_train,label_train.T),np.dot(data_train,data_train.T))
    #A=label_train.T @ data_train @ data_train.T @ label_train
    alpha=cp.Variable((N,1))
    v = np.multiply(y, X)
    print(v.T.shape)
#     obj=cp.Maximize(cp.sum(alpha)-0.5*cp.quad_form(alpha,A))
    obj=cp.Maximize(cp.sum(alpha)- 0.5 * cp.sum_squares(v.T @ alpha))
    constraints=[
        alpha <= regularisation_para_C / N,
        alpha>=0,
        cp.sum(cp.multiply(alpha,label_train))==0
    ]
    prob=cp.Problem(obj, constraints)
    prob.solve()
    svm_model={
        'alpha':alpha.value
    }
    return svm_model
```

```python
svm_model = svm_train_dual ( data_train[:LL] , label_train[:LL] , regularisation_para_C )
```

```python
def getParams(svm_model, data_train, label_train):
    N=len(label_train)  # dataset size
    alpha=svm_model['alpha']  # N
    # X data_train: NxM  # M -> feature size
    # y label_train: N
    # w=np.sum(np.dot(np.dot(data_train, alpha),label_train))
    # print(data_train.shape, label_train.shape, alpha.shape)
    # (100, 200)              (100, 1)           (100, 1)
    #    100, 200      100, 1
    # print((alpha * label_train).shape)
    w = data_train.T @ (alpha * label_train)
    # w -> Nx1
    
    index=[]
    M, _ = alpha.shape
    for i in range(M):
        if alpha[i]>0:
            index.append(i)
    b=0
    x = data_train
    for i in index:
        b+=label_train[i]
        for j in range(N):
            b-=alpha[i]*label_train[i]*np.dot(x[i],x[j])
    b/=N
    return w,b
```
